```python
R_C = ['ru 68ohms  trial 2 PEIS_C02.txt','ru 68ohms series 10muF PEIS_C02.txt']
R_CR = ['ru 68ohms series 1nF parallel 4point7kohm PEIS_C02.txt']
R_CR_CR = ['partb circuit1 PEIS_C02.txt','partb circuit2 PEIS_C02.txt','partb circuit3 PEIS_C02.txt','partb circuit6 PEIS_C02.txt']
R_CR_CR_fit = ['partb circuit1 PEIS_C02_zfit.txt','partb circuit2 PEIS_C02_zfit.txt','partb circuit3 PEIS_C02_zfit.txt','partb circuit6 PEIS_C02_zfit.txt']
```

```python
df_R_C = [pd.read_csv(file, delimiter='\t')for file in R_C]
df_R_CR = [pd.read_csv(file, delimiter='\t')for file in R_CR]
df_R_CR_CR = [pd.read_csv(file, delimiter='\t')for file in R_CR_CR]
df_R_CR_CR_fit = [pd.read_csv(file, delimiter='\t')for file in R_CR_CR_fit]
```

```python
names = ['Setup_1','Setup_2','Setup_3','Setup_4']
for n,df in enumerate(df_R_CR_CR):
    plt.clf()
    plt.figure(num=1, figsize=[10,12])
    plt.grid(linestyle = '--')
    plt.title(names[n])
    plt.xlabel('$Z_{Real}$ ($\Omega$)', fontsize=12)
    plt.ylabel('$Z_{Img}$ ($\Omega$)', fontsize=12)
    plt.plot(df['Re(Z)/Ohm'],df['-Im(Z)/Ohm'], label = 'Experimental')
    plt.plot(df_R_CR_CR_fit[n]['Re(Z)/Ohm'],df_R_CR_CR_fit[n]['-Im(Z)/Ohm'], label = 'EC-Lab Best Fit')
    plt.legend()
    plt.savefig(names[n] + '.png',bbox_inches='tight',format='png')
```

```python
warburg = ['PEIS_Fephen_1mmFe_50C_C01.txt','PEIS_Fephen_1mmFe_60C_C01.txt','PEIS_Fephen_1mmFe_70C_C01.txt','PEIS_Fephen_1mmFe_80C_C01.txt','PEIS_Fephen_1mmFe_90C_C01.txt']
warburgfit = ['PEIS_Fephen_1mmFe_50C_C01_zfit.txt','PEIS_Fephen_1mmFe_60C_C01_zfit.txt','PEIS_Fephen_1mmFe_70C_C01_zfit.txt','PEIS_Fephen_1mmFe_80C_C01_zfit.txt','PEIS_Fephen_1mmFe_90C_C01_zfit.txt']
df_warburg = [pd.read_csv(file, delimiter='\t')for file in warburg]
df_warburgfit = [pd.read_csv(file, delimiter='\t')for file in warburgfit]
```

```python
names = ['50C','60C','70C','80C','90C']
for n,df in enumerate(df_warburg):
    plt.clf()
    plt.figure(num=1, figsize=[10,12])
    plt.grid(linestyle = '--')
    plt.title(names[n])
    plt.xlabel('$Z_{Real}$ ($\Omega$)', fontsize=12)
    plt.ylabel('$Z_{Img}$ ($\Omega$)', fontsize=12)
    plt.plot(df['Re(Z)/Ohm'],df['-Im(Z)/Ohm'], label = 'Experimental')
    plt.plot(df_warburgfit[n]['Re(Z)/Ohm'],df_warburgfit[n]['-Im(Z)/Ohm'], label = 'EC-Lab Best Fit')
    plt.legend()
    plt.savefig(names[n] + '.png',bbox_inches='tight',format='png')
```

```python
names = ['50C','60C','70C','80C','90C']
plt.clf()
for n,df in enumerate(df_warburg):
    plt.figure(num=1, figsize=[10,12])
    plt.grid(linestyle = '--')
    plt.title('$Fe(phen)_3^{2+}$ Impedance at All Measured Temperatures')
    plt.xlabel('log($Z_{Real})$ ($\Omega$)', fontsize=12)
    plt.ylabel('log($Z_{Img}$) ($\Omega$)', fontsize=12)
    plt.plot(np.log10(df['Re(Z)/Ohm']),np.log10(df['-Im(Z)/Ohm']), label = names[n])
    plt.legend()
    plt.savefig('Fe(phen)3Impedance.png',bbox_inches='tight',format='png')
```
