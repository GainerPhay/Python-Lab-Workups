<h1>Introduction</h1>  
  
  &emsp; Being one of Earth's most abundant metals, iron is a very appealing target around which to design electrocatalysts and energy storage systems. In aqueous media, iron exists as an Fe<sup>2+</sup> or Fe<sup>3+</sup> cation in a 6-coordinate geometry attached to water or other ligands.<sup>1</sup> These complexes have various thermodynamic and kinetic properties based on the nature of their ligands.  This Lab aims to investigate electrochemical properties of three iron complexes, Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup>, Fe(CN)<sub>6</sub><sup>4-</sup>, and Fe(phen)<sub>3</sub><sup>2+</sup>, using potential step and potential sweep experiments.  
  
  &emsp; A system is said to be electrochemically reversible, irreversible, or quasireversible based on the kinetics of charge transfer at the working electrode.<sup>2</sup>. Systems with facile charge transfer kinetics (large rate constants) require very little energy beyond $\Delta$G<sup>0</sup> to switch between the forward and reverse reactions, and are thus considered reversible. Systems with very slow charge transfer kinetics (small rate constants) require significant energy to overcome the activation barrier between the forward and reverse reactions, and are considered irreversible. Systems with charge transfer kinetics somewhere in-between (k<sup>0</sup> ~ 10<sup>-3/-4</sup>) are considered quasireversible.  
  &emsp; The reversibility of a redox couple reaction can be evaluated by cyclic voltammetry.<sup>3</sup> Within a cyclic voltammogram, the overpotential required to drive the reaction is seen in how far from its standard reduction potential E<sup>0</sup> faradaic current starts being produced. This is most easily visualized as the distance between the anodic and cathodic current peaks. The greater the distance, the less reversible the reaction.  
  
  &emsp; Diffusive properties of a redox couple can be evaluated through potential step experiments. At overpotentials high enough to overwhelm kinetic barriers (the Cottrell region), the current response of a diffusion-controlled system to a constant applied potential varies with time according to the Cottrell equation:<sup>2</sup>  
  &emsp;
  <div align='center'> $i(t) = \dfrac{nFAD_A^{1/2}C_A^*}{\pi^{1/2}t^{1/2}}$</div>
  <div align='right'> (Eq. 1)</div>  
  where $i$ is current ($\frac{C}{s}$), $t$ is time ($s$), $n$ is the molar electron coefficient, $F$ is Faraday's constant (96,485 $\frac{C}{mol}$), $A$ is the geometric surface area of the electrode ($cm^2$), $D_A$ is the diffusion coefficient of reacting species A ($\frac{cm^2}{s}$), and $C_A^*$ is the bulk concentration of A ($\frac{mol}{L}$). Note that the geometric surface area is used over the microscopic surface area, as the system is diffusion controlled and the time it takes to diffuse from the electrode surface to a pore is infinitesimal compared to diffusion from the bulk to the surface.  
  A plot of $i$ vs $\frac{1}{t^{1/2}}$ is known as a Cottrell plot, and the diffusion coefficient can be extracted from the value of the slope of the linear best fit.
  
   

<h1>Experimental</h1>  
  
  &emsp; A Fe(CN)<sub>6</sub><sup>4-</sup> electrochemical cell was prepared in a 20 mL vial using a solution of 5.0 mM K<sub>4</sub>Fe(CN)<sub>6</sub> in 0.20 M KCL<sub>(aq)</sub>, a thoroughly polished glassy carbon working electrode, a Pt counter electrode, and a saturated calomel reference electrode (SCE). The electrodes were held in place by pushing them through a cap with cut holes, then screwing the cap onto the vial. The cell was purged with N<sub>2</sub> gas for 10 minutes before testing.
  
  &emsp; Cyclic voltammograms of the cell were taken at scan rates of 10, 20, 50, 100, 200, and 400 $\frac{mV}{s}$ across a potential window of -0.3 to 0.8 $V$ vs SCE. The CV results were used as a guide to estimate where the Cottrell region would be for the following potential step experiments.
  &emsp; Chronoamperometric scans of the cell were taken at 100, 200, 300, 650, and 750 $mV$ vs SCE, each one for 10 $s$. These potentials were chosen to start outside the Cottrell region then climb within it, and the scan time of 10 $s$ was chosen to get a significant amount of data without letting the diffusion layer grow large enough to be affected by convection in the electrolyte. Between each scan it was necessary to wait for any Fe(CN)<sub>6</sub><sup>3-</sup> produced to diffuse away from the electrode. Once the open circuit potential equilibrated, the next scan was run.  
  
  &emsp; A Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> electrochemical cell was prepared in the same 20 mL vial apparatus using a solution of 5.0 mM FeSO<sub>4</sub> in 0.20 M K<sub>2</sub>SO<sub>4</sub>. The cell was purged with N<sub>2</sub> gas for 10 minutes.
  
  &emsp; CVs of this cell were taken at the same scan rates as the Fe(CN)<sub>6</sub><sup>4-</sup> cell, this time across a potential window of -0.4 to 1.15 $V$ vs SCE. CAs were taken for 10 $s$ apiece at 600, 750, and 900 $mV$ vs SCE.
  
  &emsp; The Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> cell was turned into a Fe(phen)<sub>3</sub><sup>2+</sup> cell by replacing 2.0 mL of electrolyte with 2.0 mL of 0.20 M 1,10-phenanthroline. The reaction caused a marked color change in the solution, going from colorless to deep red.
  
  &emsp; CVs of the Fe(phen)<sub>3</sub><sup>2+</sup> cell were taken at the same scan rates as Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup>, across a potential window of 0.3 to 1.15 $V$ vs SCE. CAs were taken for 10 $s$ apiece at 800, 900, 950, and 1100 $mV$ vs SCE.

<h1>Discussion</h1>  
  
   &emsp;The end fitting of simulations to lab data was somewhat nebulus and unscientific, but the final results seem reasonable. The diffusion constants for each species are within the same magnitude as what was calculated experimentally in Table 1. Furthermore, it makes sense that Fe(CN)<sub>6</sub><sup>4-</sup> and Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> diffuse similarly as they are similar in size. Conversely, Fe(phen)<sub>3</sub><sup>2+</sup> would be expected to have a much lower diffusion coefficient due to the bulk of the 1,10-phenanthroline ligands, and that is also seen in both Table 1 and Table 2 (despite EC-Lab's software restriction). The electrode surface areas are wildly different than the geometric surface area, as the glassy carbon electrode had a measured radius of 1.5mm. Raising this value in the parameters implies theat there are deformities in the GC electrode that increase its microscopic surface area. Unlike what was mentioned regarding the Cottrell equation, microscopic surface area can have a large impact on cyclic voltammetry. Literature values for k<sup>0</sup> vary and seem to depend on pH due to H<sup>+</sup> breaking down film/residue on carbon electrodes.<sup>4</sup> As it is difficult to evaluate k<sup>0</sup> quantitatively, it is done qualitatively within the context of reversibility.  
   
   &emsp;The electrochemical reversibility judgments made in Figure 1 were based on peak separation, but they also could have been done based on how the current peak potentials shifted with scan rate. The Fe(phen)<sub>3</sub><sup>2+</sup> potentials did not change with scan rate, the hallmark of an electrochemically reversible system. The Fe(CN)<sub>6</sub><sup>4-</sup> potentials stayed fairly close, but did change with scan rate, likely quasireversible. The Fe(H<sub>2</sub>O)<sub>6</sub> potentials clearly shifted outward as scan rate increased, thus appearing irreversible.
   
  &emsp;The fitted rate constants shown in Table 2 suggest that the Fe(phen)<sub>3</sub><sup>2+</sup> cell is electrochemically reversible, the Fe(CN)<sub>6</sub><sup>4-</sup> cell is quasireversible bordering on reversible, and the Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> cell is also quasireversible, bordering on irreversible. This is a slight shift from the judgments made in Figure 1, but it follows the same trend.   
  
   &emsp;Marcus theory states that in order for a reactant species to be converted to a product species, their intermediate transition states must be isoelectronic.<sup>1</sup> For a solution of a metal-ligand complex, the energies of the complex molecules are described by a Boltzmann distribution based on the relative position of the ligands relative to the metal center. A stronger metal-ligand bond means ligands will vibrate a shorter distance to- and from- the metal center, yielding a Boltzmann distribution with fewer energy states. As a Marcus Theory transition state is one within the Boltzmann distribution, a species having fewer possible energy states is more likely to be present in one that is isoelectronic to its product species. In short, by Marcus Theory, the stronger the bond between the metal and ligand, the more favorable the kinetics.  
   
   &emsp; The results are well explained by Marcus Theory. Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> has weaker bonding than either Fe(CN)<sub>6</sub><sup>4-</sup> or Fe(phen)<sub>3</sub><sup>2+</sup> as both of the latter complexes exhibit $\pi$* -backbonding (evidenced by the coloration of the complexes). Furthermore, Fe(phen)<sub>3</sub><sup>2-</sup> has fewer possible vibrational states than Fe(CN)<sub>6</sub><sup>2-</sup> as the bidentate 1,10-phenanthroline ligands restrict the complex's freedom of movement. The predicted result based on these factors would be k<sup>0</sup><sub>Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup></sub>\<k<sup>0</sup><sub>Fe(CN)<sub>6</sub><sup>4-</sup></sub>\<k<sup>0</sup><sub>Fe(phen)<sub>3</sub><sup>2+</sup></sub>, and this is borne out by the results seen in Table 2.  
   
   &emsp; An additional explanation of the complexes' relative kinetics can be surmised based on each complex's color. That a metal complex is strongly colored indicates the presence of a t<sub>2g</sub> &rarr; $\pi$* electronic transition.<sup>4</sup> This transition is the energetic pathway by which an electron leaves when Fe<sup>2+</sup> oxidizes to Fe<sup>3+</sup>. The Fe(CN)<sub>6</sub><sup>4-</sup> solution is yellow, and thus absorbs violet light. The Fe(phen)<sub>3</sub><sup>2+</sup> solution is red, and thus absorbs green light. As green light is lower energy than violet light, the t<sub>2g</sub> &rarr; $\pi$* transition (and thus oxidation) of Fe(phen)<sub>3</sub><sup>2+</sup> is more favorable than that of Fe(CN)<sub>6</sub><sup>4-</sup>. Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> is colorless, indicating it has no t<sub>2g</sub> &rarr; $\pi$* transition, and so its electrons leave through a t<sub>2g</sub> &rarr; e<sub>g</sub> transition. This is a Laporte-forbidden d &rarr; d transition, so the Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> complex must distort its symmetry before it can oxidize. Due to this extra energy barrier, it reasons that Fe(H<sub>2</sub>O)<sub>6</sub><sup>2+</sup> would show slower kinetics than either Fe(CN)<sub>6</sub><sup>4-</sup> or Fe(phen)<sub>3</sub><sup>2+</sup>, which again is borne out by Table 2.  
   
   &emsp; One final note that the conclusions drawn from the data address electrochemical reversibility, not chemical reversibility. Electrochemical reversibility speaks specifically to the interactions between species and an electrode, and it indicates the kinetic limitations needed to be overcome to reverse a reaction. Chemical reversibility whether a reaction can be reversed at all. No species studied in this Lab left their cell systems, nor was their oxidation so favorible that a subsequent reduction back became thermodynamically impossible. Thus, every system studied showed chemical, but not necessarily electrochemical reversibility.
