<h1><u>Introduction</u></h1>  

The Hydrogen Evolution Reaction (HER) and Oxygen Evolution Reaction (OER) are centerpieces of fuel cell technology, and making these reactions more efficient is a continual goal of electrochemists. Both HER and OER are kinetically hindered by charge transfer, requiring significant overpotentials to drive the reaction.<sup>(1)</sup> Many electrocatalysts are used to bring this energy barrier down, and three of them will be investigated in this lab.
<br>
HER and OER can be catalyzed in both acidic and basic conditions. In acid, they follow Eq. 1 and Eq. 2 respectively:

Thermodynamically, HER is easiest in acid, and OER is easiest in base, so catalysts are often designed around those conditions.


Electrochemical kinetics is studied by Tafel analysis of cyclic voltammograms. Regions in a CV where current varies exponentially with potential show where charge transfer kinetics is dominant. By taking the log of the currents, the exponential region becomes a much easier to analyze linear region. The linear best fit equation y = mx + b gives two important values: m = the Tafel slope (how much potential does it take to drive the reaction 10-fold) and b = the exchange current (how much current is running back and forth through the electrode at equilibrium).

In this lab, Pt<sup>0</sup> and Cu<sup>0</sup> are used for acid catalysis of HER, and Fe-doped NiOOH is used for base catalysis of OER. Tafel analysis is performed on the resulting data to obtain kinetic information that can be compared to known literature values.

<h1><u>Methods</u></h1>  

The following working electrodes were studied for HER catalysis: A Pt electrode cleaned by electric current pulsing, a Pt electrode cleaned by flame annealing, a Pt electrode that had been plated with Pt-Black after cleaning, and Cu wire.
For OER catalysis a Fe-doped nickel oxide Fe-NiOOH is deposited onto a gold electrode.

<u><i>All Reactions</i></u>  
All experiments were done via potentiostat in a three-electrode cell using a Pt counter electrode and an Ag/AgCl reference electrode. In particular, the reference used in each experiment was a Ag/AgCl electrode in saturated KCl that measured at -0.30 V vs. a Saturated Calomel Electrode (-0.211 V vs. SHE). Each cell used a Teflon stir bar and was purged with N<sub>2</sub> gas for 10 minutes. The software used to control the potentiostat and integrate peaks was EC-Lab.
Before running a potential sweep with EC-Lab, the uncompensated series resistance of the electrolyte was measured using impedance spectroscopy.
    
<u><i>Pt Catalyzed Hydrogen Evolution</i></u>  
A three-neck round-bottom flask was filled with 20 mL 1.0 M H<sub>2</sub>SO<sub>4</sub>(aq). After purging with N<sub>2</sub> gas, the N<sub>2</sub> stream is moved out of the liquid so that it flows over the top of the acid solution. It's important for the air in the cell to be kept as only N<sub>2</sub>, as any H<sub>2</sub> or O<sub>2</sub> gas present would change the equilibrium of the cell.
A used Pt electrode was electrically cleaned by connecting it to the potentiostat and running pulses of -5.0 and +5.0 mA current into the electrode, alternating every second for 10 seconds. This electrode is placed in the cell alongside the reference and counter electrodes described above. 
Before running HER, it's important to determine the active surface area of the electrode. To do this, sweeps were run (unstirred) from 0.0 V to 1.2 V vs RHE at scan rates of 20 mV/s, 100 mV/s, and 200 mV/s, being careful not to produce H<sub>2</sub> gas. The CV peaks that result come from the formation of Pt oxides at high potential and the adsorption of hydrogen at low potential. After a scan finished, the area under the hydrogen adsorption curve was calculated using EC-Lab, giving a charge Q (mC) as a result.
To run HER, the stir bar was turned on and the cell was swept from -0.3 V to 0.5 V vs RHE at a scan rate of 10 mV/s. This result is what's used for Tafel analysis.
Following the sweep, the electrode was removed and subjected to flame annealment to remove any impurities. It was then placed back in the cell and tested with the same two potential sweeps as listed above. Afterwards, Pt-Black was deposited onto the working electrode using an electroplating cell of 1% wt H<sub>2</sub>PtCl<sub>6</sub> and 0.02% wt PbNO<sub>3</sub> set to -2 V at the working electrode. After 3 minutes, the now Pt-Black electrode was removed and tested with the same potential sweeps above.
    
<u><i>Cu Catalyzed Hydrogen Evolution</i></u>  
    A Cu wire electrode was tested through the same HER procedure as the Pt electrodes above.
    
<u><i>Fe-NiOOH Catalyzed Oxygen Evolution</i></u>  
    A three-neck round-bottom flask was filled with 20 mL 1.0 M KOH(aq). The system was purged with N<sub>2</sub> gas in the same manner as the HER prodecure above.
    A commercial Au/Ti electrode was placed in a two-electrode cell containing a deposition solution of 0.10 M Ni(NO<sub>3</sub>)<sub>2</sub>(aq) and 0.003 M FeCl<sub>2</sub>(aq). A cathodic current of 1 mA/cm<sup>2</sup> was run using chronopotentiometry in EC-Lab for 10 seconds.
    The plated electrode was then placed in the three-neck cell with the counter and reference electrodes. A potential sweep from 0.24 V - 0.48 V was run (no-stir)at 20 mV/s and 50 mV/s to obtain a Ni(II) oxidation peak to be used in active site calculations. After this, full CVs with stirring were taken from 0.0 V to 1.0 V at both 20 mV/s and 50 mV/s.

<h1><u>Discussion</u></h1>  

The surface area based experiment among the Pt electrodes gives an unsurprising result: with more catalyst, kinetics are better. The plated electrode had a microscopic surface area much larger than its geometric area, and so it had many more Pt sites at which H<sup>+</sup> could be adsorbed.

The Tafel plot of Cu is very suspicious in that it has a substantial positive, measured current at overpotential 0.0. By theory, log(i) should go to -infinity at $\eta$ = 0.0. This error may be explained by a degradation of the Ag/AgCl reference electrode by the end of the first day of data collection. As the reference was returned to saturated KCl for a week before running the OER experiment, it is possible that the electrode re-equilibrated to standard condition before Fe-NiOOH data was measured. Perhaps more likely is an error in the measurement of uncompensated series resistance leading to an overcorrection of potential.
Should the correct Cu Tafel plot shift positively along the x-axis, the y-intercept would occur at a lower point, yielding an exchange current closer in magnitude to those of Platinum. The Tafel slope should remain the same, and it being much larger indicates that charge transfer is the dominant kinetic term over a wider range of potentials.

The exchange current for the OER is remarkably small, but this is justified as the oxidation occurred at a significant overpotential. Both a high overpotential requirement and a low exchange current indicate slow kinetics.
The Tafel slopes in Figure 7 are very close to the upper end of the literature range: 37 $\pm$ 3 mA/decade.<sup>(3)</sup> They do not appear to change with voltage, but one would expect at higher scan rates it would. At a higher scan rate, the Ni(II) oxidation peak would broaden to the point where some portion of Ni(II) oxidation would occur at the same time as OER. Then as potential increases, more Ni(III) active sites are formed, and the Tafel slope declines.
The turnover frequency calculated also aligns well with literature, falling within the range [0.03-0.08 s<sup>-1</sup>]<sup>(3)</sup>
