<h2><u>Introduction</u></h2>  
  
  Electrochemical cells are systems that have current run through various impedances, and as such, they can be modeled using circuits. Impedance spectroscopy is an analytical technique that takes advantage of this by measuring how current and potential interact upon application of Alternating Current. As capacitors, resistors, and inductors have known response patterns to AC, the results of impedence spectroscopy can be used to create a quantitative circuit model of an analyzed cell, with each part of the circuit corresponding to some electrochemical phenomenon.
  The simplest circuit model of an electrochemical cell is known as a Randles circuit, depicted below:  
  
 ![](randles.svg)  
  
  <div align='center'> <b>Figure 1:</b> A basic Randles circuit</div>  
  
  Here, R<sub>s</sub> corresponds to the the series resistance of current moving through an electrolyte, C<sub>DL</sub> corresponds to the capacitance of the electric double layer that forms at the working electrode,and R<sub>CT</sub> corresponds to the kinetic energy barriers that must be overcome to transfer electrons between chemical species and the electrode. The impedance of this circuit to an AC of frequency $\omega$ is described by the following equations:<sup>1</sup>  
  &emsp;
  &emsp;
  <div align='center'>$Z_{real} = R_S + \dfrac{R_{CT}}{1 + \omega^2R_{CT}^2C_{DL}^2}$ &emsp; &emsp; $Z_{img} = \dfrac{R_{CT}^2C_{DL}\omega}{1 + \omega^2R_{CT}^2C_{DL}^2}$</div>
  <div align='right'>(Eq. 1,2)</div>  
  Impedance tends to be handled in the complex plane as a convenient way to keep track of the response time delay (phase shift) between voltage and current. The real component of $Z$ describes the resistive effects in the cell, while the imaginary component describes the capacitative effects. 
  Figure 1 best corresponds to a three-electrode cell where impedance effects at the counter electrode can be ignored since the working electrode is being measured against a reference, not the counter. For two-electode cells, another RC component is added to the model to account for the physics at the counter electrode:  
  
  ![](two_randles.svg)
  
<div align='center'><b>Figure 2:</b> Randles circuit with two electrodes</div>

Using the Randles circuit shown in Figure 1 as a base, the impedances from additional circuit components contribute directly to the total impedance as described below:<sup>1</sup>  
  <div align='center'>$Z_{total} = Z_1 + Z_2 + Z_3 + ...$</div> 
  <div align='right'>(Eq. 3)</div>  
  For diffusion-controlled electrochemical reactions, the circuit model for the working electrode is shown below:  
    
   ![](warburg_circuit.svg)  
    
  <div align='center'><b>Figure 3:</b> Randles circuit with Warburg impedance</div>  
  
$Z_W$ refers to Warburg impedance, the effective reduction of current response due to diffusion limiting how much reactant reaches the electrode. 
The following Lab has two goals: 1) Demonstrate the efficacy of impedance spectroscopy through use on model Randles circuits; 2) Use electrochemical impedence spectroscopy to investigate the temperature dependence of Warburg impedance within a three-electrode Fe(phen)<sub>3</sub><sup>2+/3+</sup> cell.

<h5>Part 2: Impedance Spectroscopy of Fe(phen)<sub>3</sub><sup>2+</sup> at Varying Temperatures</h5>  
    
   Fe(phen)<sub>3</sub><sup>2+/3+</sup> was chosen for study as in previous labs it has shown to be an electrochemically reversible cell, so there should be little overlap between when the reaction is kinetics-dominated and when its diffusion-dominated. This allows diffusion effects such as Warburg Impedance to be better isolated.  
    
   A three-electrode Fe(phen)<sub>3</sub><sup>2+/3+</sup> cell was created in a 20 mL glass vial by adding 1.0 mL of 200 mM FeSO<sub>4</sub> to 19.0 mL of 0.1M K<sub>2</sub>SO<sub>4</sub>, then replacing 3 mL of this solution with 3 mL of 1,10-phenanthroline (in sulfate buffer). A Glassy Carbon Electrode was used as the working electrode, a platinum wire was used as the counter electrode, and a Saturated Calomel Electrode was used as the reference. These electrodes were placed in holes in the vial's cap, which was screwed on after the solution had been purged with N<sub>2</sub> gas for 10 minutes. The cell was then connected to a potentiostat controlled by EC-Lab.  
    
   This cell was held in a water bath atop a hot plate, and the temperature of the bath was monitored by a thermometer. Impedance measurements scanning frequencies from 1MHz to 10 Hz were taken when the temperature of the bath hit 50 °C, 60 °C, 70 °C, 80 °C, and 90 °C.

<h2><u>Discussion</u></h2>  
   Figure 5a illustrates the separation of a circuit's series resistance impedance and its capacitance-contributed impedance. The impedance from series resistance stays constant at 68 $\Omega$, while the capactive impedance moves solely in the Z<sub>Img</sub> direction. This is consistent with Eq. 1 and Eq. 2 as shown earlier.  
   
   The characteristic shape of a Nyquist plot of an R + R/C circuit is a semicircle as seen in Figure 5b. The left endpoint matches the series resistance R<sub>S</sub> = 68 $\Omega$, and the right endpoint matches the total resistance R<sub>Total</sub> = R<sub>S</sub> + R<sub>CT</sub>. This is consistent with Eq.1 as when $\omega$ gets very large, the R<sub>CT</sub> term should approach a value of R<sub>CT</sub>.  
   
   The fit data in Table 3 aligns well with stated component values listed in Table 2 for the first and second study, but they wildly diverge for studies 3 and 4. Judging by the graphs in Figure 6, it appears that the better separated the semicircles corresponding to circuit elements, the better the fit gets. This makes sense as when there is complete overlap, it is impossible to statistically distinguish the contributions of any one component to the total data. It appears the main factors leading to model failure are using capacitors of equal value or using resistors of disparate values. In a circuit of the form shown by Figure 4, having equal capacitors results in current charging both at the same time, so their characteristic Nyquist peaks occur at the same point and are thus indistinguishable. In the case of having resistors of different magnitudes, the degrees to which the capacitors charge before current flows through the resistor is wildly different, leading again to an incomplete picture of capacitance.
   
   A log graph was chosen to depict the differences between the impedance spectra across temperatures because they are very small and hard to see on a standard linear graph. Figure 7a clearly shows that the impedances differ with temperature, and they follow a trend of decreasing with increasing temperature.
   
   A R1 + C2/(R2 + W2) fit was chosen because that is what is decribed by theory (Figure 3). While other fits matched the data better, they offer no additional value without an accompanying physical explanation. The fit data in Table 4 shows two trends: the capacitance of the cell decreases with temperature and the Warburg impedance rises with temperature.
   It makes sense that the Warburg impedence comes out with a $t^{-\frac{1}{2}}$ term, as limiting diffusion results in a concentration gradient that changes with $t^{-\frac{1}{2}}$.  
    
   The trend of Warburg resistance versus temperature (Figure 8) is unexpected. It reasons that at higher temperature would increase the overall ion movement in solution and thus increase the rate of diffusion. This would lessen the diffusion limitation on current as more species would get to the electrode to generate current per unit time. It is unclear whether the trend seen is due to a yet unconsidered physical aspect or from issues with fit resolution similar to that seen in Part 1.  
   
   As current affected by diffusion shows a dependance on time, it reasons that the effect on impedence would show up in the imaginary component of Z. Therefore, it is possible that the curve fits mixed up effects caused by capacitance and those caused by Warburg impedance. Both capacitance and Warburg impedance showed trends across temperature, but perhaps only one should have shown a trend that essentially combines the two trends.
